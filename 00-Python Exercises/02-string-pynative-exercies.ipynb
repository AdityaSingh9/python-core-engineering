Given string contains a combination of the lower and upper case letters. Write a program to arrange the characters of a string so that all lowercase letters should come first.

Given:

str1 = PyNaTive

Expected Output:

yaivePNT


```python
str1 = "PyNaTive"
ls1=[]
ls2=[]
for x in str1:
    if x == x.upper():
        ls2.append(x)
    else:
        ls1.append(x)
print("".join(ls1+ls2))
```

Count all letters, digits, and special symbols from a given string
Given:

str1 = "P@#yn26at^&i5ve"
Expected Outcome:

Total counts of chars, digits, and symbols 

Chars = 8 
Digits = 3 
Symbol = 4

```python
def countEntity(str1):
    Chars = 0
    Digits = 0 
    Symbol = 0
    for x in str1:
        if x.isnumeric():
            Digits +=1
        elif x.isalpha():
            Chars +=1
        else:
            Symbol +=1
    print(Chars , Digits , Symbol)

countEntity("P@#yn26at^&i5ve")
```

Write a program to count occurrences of all characters within a string

Given:

str1 = "Apple"

Expected Outcome:

{'A': 1, 'p': 2, 'l': 1, 'e': 1}

```python
str1 = "Apple"
myDict={}
for x in str1:
    if x in myDict:
        myDict[x]=myDict[x]+1
    else:
        myDict[x]=1

print(myDict)
```
